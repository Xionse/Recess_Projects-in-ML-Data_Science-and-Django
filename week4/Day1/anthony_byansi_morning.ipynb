# Data Preprocessing


1. **Data Cleaning** :
   - Handling Missing Values :
     - `df.dropna()` : Drops rows or columns with missing values.
     - `df.fillna()` : Fills missing values with specified methods like mean, median, or forward/backward fill.
   - Outlier Detection and Treatment :
     - `z-score` or `IQR` : Identifies outliers based on statistical measures and removes or transforms them.

2. **Data Transformation** :
   - Feature Scaling :
     - `StandardScaler()` : Scales numerical features to have zero mean and unit variance.
     - `MinMaxScaler()` : Transforms features to a specified range, typically between 0 and 1.
   - Handling Skewed Data :
     - `np.log()` : Applies a logarithmic transformation to reduce skewness.
     - `PowerTransformer()` : Performs power transformations like Box-Cox or Yeo-Johnson to reduce skewness.

3. **Feature Engineering** :
   - Polynomial Features :
     - `PolynomialFeatures()` : Generates polynomial features by combining existing features up to a specified degree.
   - Feature Selection :
     - `SelectKBest()` : Selects the top K features based on statistical measures like chi-square or mutual information.
     - `Recursive Feature Elimination (RFE)` : Selects features recursively based on model performance.

4. **Handling Categorical Data** :
   - One-Hot Encoding :
     - `pd.get_dummies()` : Converts categorical variables into binary one-hot encoded vectors.
   - Target Encoding :
     - `category_encoders.TargetEncoder()` : Replaces categorical variables with the mean of the target variable within each category.

5. **Dimensionality Reduction** :
   - Principal Component Analysis (PCA) :
     - `sklearn.decomposition.PCA()` : Performs dimensionality reduction by projecting data onto principal components.
   - t-SNE :
     - `sklearn.manifold.TSNE()` : Reduces high-dimensional data into a lower-dimensional space for visualization or clustering.

6. **Data Sampling** :
   - Random Sampling :
     - `df.sample()` : Selects a random sample from the dataset.
   - Stratified Sampling :
     - `sklearn.model_selection.StratifiedShuffleSplit()` : Performs stratified sampling based on class labels.

7. **Time Series Data** :
   - Resampling :
     - `df.resample()` : Aggregates or interpolates time series data into different time intervals.
   - Lagging :
     - `shift()` : Creates lag features by shifting the time series data.

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/content/bse_dataset4.csv')

# Iterate over each column in the DataFrame
for column in df.columns:
    #appropriate visualization based on column type
    if df[column].dtype == 'object':  # Categorical column
        plt.figure(figsize=(8, 6))
        sns.countplot(data=df, x=column)
        plt.title(f'{column} Distribution')
        plt.xticks(rotation=45)
        plt.show()
    elif df[column].dtype == 'int64' or df[column].dtype == 'float64':  # Numeric column
        plt.figure(figsize=(8, 6))
        sns.histplot(data=df, x=column, kde=True)
        plt.title(f'{column} Distribution')
        plt.show()
    else:  # Other column typ

        pass
````

````python

# Load the dataset
df = pd.read_csv('/content/bse_dataset4.csv')

# Iterate over each column in the DataFrame
for column in df.columns:
    fig, ax = plt.subplots()
    if df[column].dtype == 'object':
        # Categorical column: Create a bar plot
        value_counts = df[column].value_counts()
        value_counts.plot(kind='bar', ax=ax)
    else:
        # Numeric column: Create a histogram
        df[column].plot(kind='hist', ax=ax)

    # Setting the title and labels
    ax.set_title(f"{column} Distribution")
    ax.set_xlabel(column)
    ax.set_ylabel("Frequency")

    # Display the plot
    plt.show()
````
